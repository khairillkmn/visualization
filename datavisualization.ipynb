```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
data = pd.read_csv('Data.csv')
data.head()
```

```python
data["Purchased"].value_counts()
```

```python
data.plot(kind="scatter", x="Age", y="Salary")
```

```python
sns.jointplot(x="Age", y="Salary", data=data, size=5)
```

```python
sns.FacetGrid(data, hue="Purchased").map(plt.scatter, "Age", "Salary").add_legend()
```

```python
sns.pairplot(data.drop("Country", axis=1), hue="Purchased", size=3, diag_kind="kde")
```
